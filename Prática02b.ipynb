```python
import pandas as pd
desp21 = pd.read_csv('desp21.csv', sep = ';')
desp21.head(5)
```

```python
desp21.drop(['ano_exercicio', 'ds_orgao', 'ds_municipio', 'mes_referencia', 'dt_emissao_despesa', 'ds_programa', 
             'ds_fonte_recurso', 'ds_cd_aplicacao_fixo', 'nr_empenho', 'cd_programa', 'cd_acao'], axis = 1, inplace = True)

#se eu não especificar o axis, ele vai tentar eliminar da LINHA

desp21.head()
```

```python
s = pd.to_numeric((desp21['vl_despesa']).str.replace(',', '.'))
```

```python
desp21['vl_despesa'] = s
```

```python
# desp21.dtypes
```

```python
tp_despesas = list(desp21['tp_despesa'].drop_duplicates())
tp_despesas
```

```python
func_governo = list(desp21['ds_funcao_governo'].drop_duplicates())
func_governo
```

```python
valores_pg = ['Valor Pago']
valores_pg
```

```python
selecao = desp21['tp_despesa'].isin(valores_pg)
selecao
```

```python
desp21_pg = desp21[selecao]

# para mudar os ÍNDICES desta série
desp21_pg.index = range(desp21_pg.shape[0])
desp21_pg
```

```python
func_gov_vl_pg = desp21_pg.groupby(["ds_funcao_governo"])['vl_despesa'].sum()
```
